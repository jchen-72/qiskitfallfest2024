# PART 0: SETUP

*This part sets up qiskit and the rounds, make sure to run it*

**Imports**

```python
from qiskit import *
import RoundList
from qiskit.visualization import plot_bloch_multivector
from qiskit.visualization import plot_histogram
from qiskit_aer import AerSimulator
from qiskit_ibm_runtime import QiskitRuntimeService
from qiskit.quantum_info import Statevector
```

**Functions for viewing state**

```python
def testBloch(circuit):
    state = Statevector(circuit)
    return plot_bloch_multivector(state)
```

```python
def testMeasure(circuit):
    circ=circuit.copy()
    circ.measure_all()
    aersim = AerSimulator()
    result_ideal = aersim.run(circ, shots=1024).result()
    counts_ideal = result_ideal.get_counts()
    return plot_histogram(counts_ideal, title='Result')
```

```python
def testDraw(circuit):
    return circuit.draw(initial_state=True, output="mpl")
```
